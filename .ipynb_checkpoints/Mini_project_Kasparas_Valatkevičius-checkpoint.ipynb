## Train/Test split

```python
mask = np.random.random(len(Banking)) < 0.8
```

## Modelling

```python
def train_and_predict(Banking, model):
    X_train = Banking[mask].drop(['y'],axis=1)
    y_train = Banking[mask]['y']
    X_test = Banking[~mask].drop(['y'],axis=1)
    y_test = Banking[~mask]['y']

    X_train = (X_train - X_train.mean(axis=0)) / X_train.std(axis=0) # Standard scaling
    X_test = (X_test - X_test.mean(axis=0)) / X_test.std(axis=0) # Standard scaling
    
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)
    
    
    print('Correctly identified on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    (pred_train == y_train).mean(),
    (pred_test == y_test).mean())),
    print('Precision on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    precision_score(y_train,pred_train),
    precision_score(y_test,pred_test))),
    print('Recall on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    recall_score(y_train,pred_train),
    recall_score(y_test,pred_test))),
    return pd.DataFrame(confusion_matrix(y_test, pred_test)).style.background_gradient()
```

```python
lr = LogisticRegression(solver='lbfgs', max_iter=2000)
dt = DecisionTreeClassifier(max_depth=50)
rf = RandomForestClassifier(2000,max_depth = 20, criterion='entropy')
```

```python
train_and_predict(Banking, lr)
```

```python
lr_coefs = lr.coef_ # We introduced standard scaling, therefore the biggest coefficient should be the most significant in logistic regression
```

```python
train_and_predict(Banking, dt)
```

```python
train_and_predict(Banking, rf)
```

```python
rf.feature_importances_ #We introduced standard scaling, therefore the biggest coefficient should be the most significant in random forest classifier
```

# Is the accuracy of our model the main criterion for our modelling success?

## No.

Remember, we are looking to identify people (lines) who are going to make a term deposit. We are trying to identify as many actual positives in the target class. We are looking for as many TRUE POSITIVES as possible and we are looking to make as few type ii (FALSE NEGATIVE) mistakes as possible. Therefore, RECALL is our main thing to look at when modelling. We must tweak our models to improve our poor recall result.

To improve our recall result we have to notice that our dataset is actually an imbalanced one. Meaning, that the two classes in the target variable have a big difference in size.\
To fight this imbalance, we must either undersample or oversample. 

```python
Banking.shape
```

We will attempt the random under sampler for our case.

```python
undersample = RandomUnderSampler()
```

```python
X, y = Banking.drop(['y'],axis = 1),Banking['y']
```

```python
X.shape
```

```python
X_res, y_res = undersample.fit_resample(X,y)
```

```python
X_res.shape #We can see a significant decrease in sample size. Let's check if it helps our case with modelling.
```

```python
y_res.shape
```

```python
Banking_res = X_res
Banking_res['y']= y_res
```

#### New mask and model for the reshaped dataframe

```python
mask_res = np.random.random(len(Banking_res)) < 0.8
```

```python
def train_and_predict_undersample(Banking_res, model):
    X_train = Banking_res[mask_res].drop(['y'],axis=1)
    y_train = Banking_res[mask_res]['y']
    X_test = Banking_res[~mask_res].drop(['y'],axis=1)
    y_test = Banking_res[~mask_res]['y']

    X_train = (X_train - X_train.mean(axis=0)) # scaling
    X_test = (X_test - X_test.mean(axis=0)) # scaling
    
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)
    
    
    print('Correctly identified on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    (pred_train == y_train).mean(),
    (pred_test == y_test).mean())),
    print('Precision on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    precision_score(y_train,pred_train),
    precision_score(y_test,pred_test))),
    print('Recall on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    recall_score(y_train,pred_train),
    recall_score(y_test,pred_test))),
    return pd.DataFrame(confusion_matrix(y_test, pred_test)).style.background_gradient()
```

```python
train_and_predict_undersample(Banking_res, lr)
```

We can see that the percentage of correct guesses overall has gone down, BUT recall has gone up. And in this case it is indeed what matters most. Let's attempt the other models and see wether recall goes up for all.

```python
train_and_predict_undersample(Banking_res, dt)
```

```python
train_and_predict_undersample(Banking_res, rf)
```

## We are losing percentage on correct guesses, but we are improving our model

Once again, it is important that we realize, that what matters in our model is not the overall percentage of correct classification, but the the size of TRUE POSITIVE to FALSE NEGATIVE ratio. In this instance we have managed to improve recall. Let's try oversampling as another option to improve recall.

```python
oversample = RandomOverSampler()

X, y = Banking.drop(['y'],axis = 1),Banking['y']

X_over, y_over = oversample.fit_resample(X,y)
```

```python
X_over.shape # In this case we see an increase in sample size. 
```

```python
Banking_over = X_over
Banking_over['y']= y_over
```

#### New mask and model for oversampled dataframe

```python
mask_over = np.random.random(len(Banking_over)) < 0.8
```

```python
def train_and_predict_oversample(Banking_res, model):
    X_train = Banking_over[mask_over].drop(['y'],axis=1)
    y_train = Banking_over[mask_over]['y']
    X_test = Banking_over[~mask_over].drop(['y'],axis=1)
    y_test = Banking_over[~mask_over]['y']

    X_train = (X_train - X_train.mean(axis=0)) / X_train.std(axis=0) # Standard scaling
    X_test = (X_test - X_test.mean(axis=0)) / X_test.std(axis=0) # Standard scaling
    
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)
    
    
    print('Correctly identified on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    (pred_train == y_train).mean(),
    (pred_test == y_test).mean())),
    print('Precision on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    precision_score(y_train,pred_train),
    precision_score(y_test,pred_test))),
    print('Recall on train set - {0:.02%}, on test set - {1:.02%}\n'.format(
    recall_score(y_train,pred_train),
    recall_score(y_test,pred_test))),
    return pd.DataFrame(confusion_matrix(y_test, pred_test)).style.background_gradient()
```

```python
train_and_predict_oversample(Banking_over, lr)
```

```python
train_and_predict_oversample(Banking_over, dt)
```

```python
train_and_predict_oversample(Banking_over, rf)
```

Incredibly, we were able to increase the recall manyfold by oversampling. Perhaps the solution to get the best possible model for our task would be to combine those two - undersampling and oversampling to get an improved method.
